```python
import csv
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
GA = nx.read_edgelist('edges.csv',delimiter=",")
```

```python
print(nx.info(GA))
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
layout = nx.circular_layout(GA)
nx.draw_networkx_nodes(GA, pos=layout, node_size=100, alpha=0.8) 
nx.draw_networkx_edges(GA, pos=layout, width=3, style='dotted',
                      edge_color='orange')
#nx.draw_networkx_labels(GA, pos=layout, font_size=15)
plt.show()
```

```python
# Degree of everyone
#GA.degree()
GA.degree("1")
```

```python
sorted(GA.degree().items(), key=lambda x:x[1], reverse=True)[:5]
```

```python
degree_nodeOne = GA.degree('1')
total_nodes_minus_nodeOne = len(GA.nodes()) - 1
```

```python
# Degree centrality for nodeOne
degree_centrality_nodeOne = (degree_nodeOne / total_nodes_minus_nodeOne)
print("Calculated degree centrality for NodeOne:", degree_centrality_nodeOne)
```

```python
# Double check using networkx built function
print("Networkx degree centrality for NodeOne:", nx.degree_centrality(GA)["1"])
```

```python
# degree centrality for all Nodes
degree_centrality = nx.degree_centrality(GA)
degree_centrality
```

```python
sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
```

```python
nx.set_node_attributes(GA, 'degree centrality', degree_centrality)
```
